```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
y = df.Survived
X = df.drop('Survived',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
y_train.value_counts(normalize=True)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=10)
```

```python
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```

```python
train_accuracy = []
validation_accuracy = []
for i in range(1,10):
    model = DecisionTreeClassifier(max_depth=i,random_state=10)
    model.fit(X_train,y_train)
    train_accuracy.append(model.score(X_train,y_train))
    validation_accuracy.append(model.score(X_test,y_test))
```

```python
result = pd.DataFrame({'depth':range(1,10),'train_acc':train_accuracy,'valid_acc':validation_accuracy})
result.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(result.depth,result.train_acc,marker='o')
plt.plot(result.depth,result.valid_acc,marker='o')
plt.xlabel('Depth of the Tree')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
train_accuracy = []
validation_accuracy = []
for i in range(1,50,5):
    model = DecisionTreeClassifier(min_samples_leaf=i,random_state=10)
    model.fit(X_train,y_train)
    train_accuracy.append(model.score(X_train,y_train))
    validation_accuracy.append(model.score(X_test,y_test))
result = pd.DataFrame({'min_samples_leaf':range(1,50,5),'train_acc':train_accuracy,'valid_acc':validation_accuracy})
result.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(result.min_samples_leaf,result.train_acc,marker='o')
plt.plot(result.min_samples_leaf,result.valid_acc,marker='o')
plt.xlabel('Min Samples Leaf')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
final_model = DecisionTreeClassifier(max_depth=5,max_leaf_nodes=20,random_state=10)
final_model.fit(X_train,y_train)
print(final_model.score(X_train,y_train))
print(final_model.score(X_test,y_test))
```

```python
from sklearn import tree
```

```python
decision_tree = tree.export_graphviz(
    final_model,out_file='tree.dot',feature_names=X_train.columns,max_depth=2,filled=True)
```
